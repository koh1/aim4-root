```python
def load_file(file_name):
    f = open(file_name, 'r')
    array_r = []
    array_t = []
    for r in f:
        line = r.strip().split(',')
        if line[2].strip() == "RESERVATION_CONFIRMED":
            array_r.append([float(line[0].strip()), int(line[1].strip()), int(line[3].strip()), int(line[4].strip())])
        elif line[2].strip() == "VEHICLE_REMOVED":
            array_t.append([float(line[0].strip()), int(line[1].strip()), float(line[3].strip())])
    df_r = pd.DataFrame(array_r, columns=["time", "vid", "conflicts", "unacceptables"])
    df_t = pd.DataFrame(array_t, columns=["time", "vid", "duration"])
    return (df_r, df_t)
```

* Intersection Managerからの応答時間を１秒程度に見積もっておくと遅延が大きくゆらいでも、Unacceptableなものを受け取ることがない。その結果、通信遅延無しの場合と比較して、交差点における車両スループット2.52 -> 2.45程度の劣化にとどまる。さらに交差点通過にかかる時間は...(計測中)
* 応答時間をどの程度の大きさに設定するべきかに関して。１秒を越えるとスループットが低下する。
* 結論として **各Vehicleが自身の通信遅延に応じてIMからの応答時間見積もりをし、それを考慮したリソース要求を実施する** ことが、交差点スループット最大化につながる
* 車両スピードなどの観点から応答時間に関する制約が異なる可能性
    * 現行は90km/h程度 (速い!)
    * 54km/hだと -> 同じく1sec -> 2secで性能劣化
